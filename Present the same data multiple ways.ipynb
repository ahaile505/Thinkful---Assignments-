```python
import pandas as pd
import seaborn as sns
%matplotlib inline
```

```python
df=pd.read_csv('purchases.csv')
```

Choose one variable and plot that variable four different ways.

```python
df.head()
```

```python
sns.distplot(df['ad_views'],kde=False,bins=10)
```

```python
sns.rugplot(df['ad_views'])
```

```python
sns.barplot(y=df['ad_views'])
```

```python
sns.countplot(x='items_purchased',data=df)
```

Choose two continuous variables, and plot them three different ways.

```python
sns.lineplot(x='items_purchased', y='ad_views', data=df)
```

```python
sns.lmplot(x='items_purchased', y='ad_views', data=df)
```

Choose one continuous variable and one categorical variable, and plot them six different ways.

```python
sns.barplot(y='ad_views',x='country',data=df)
```

```python
sns.boxplot(y='ad_views',x='country',data=df)
```

```python
sns.violinplot(y='ad_views',x='country',data=df)
```

```python
sns.stripplot(y='ad_views',x='country',data=df)
```

```python
sns.pointplot(y='ad_views',x='country',data=df)
```

```python
sns.swarmplot(y='ad_views',x='country',data=df)
```
